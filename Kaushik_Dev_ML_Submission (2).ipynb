Project Summary: Machine Failure Prediction using Autoencoder-Based Anomaly Detection

Introduction

Predictive maintenance is a crucial aspect of industrial operations, helping businesses reduce downtime, optimize resource utilization, and extend the lifespan of machinery. This project focuses on developing an anomaly detection model using an autoencoder-based neural network to predict machine failures. The dataset contains sensor readings and machine operating conditions, with labeled instances indicating whether a machine has failed.

Objective

The primary objective of this project is to develop a machine learning model that can detect potential machine failures based on sensor data. By leveraging an autoencoder, the model aims to identify anomalous behavior in machines before failures occur. This helps businesses shift from reactive maintenance (fixing after failure) to predictive maintenance (fixing before failure), reducing unexpected breakdowns and associated costs.

Dataset Overview

The dataset consists of various numerical sensor readings collected from industrial machines. The key features include:

Torque – Measures the force applied to the rotating parts.

Rotational Speed – Indicates how fast the machine components are moving.

Air Temperature – Monitors external environmental conditions.

Process Temperature – Tracks internal heat generation during operation.

Power Consumption – Reflects the energy used by the machine.

Machine Failure (Target Variable) – Binary label indicating machine failure (1) or normal operation (0).

Model Selection and Justification

We implemented an autoencoder, a type of unsupervised neural network, for anomaly detection. The autoencoder learns to reconstruct normal machine behavior and detects failures by identifying instances with high reconstruction errors. The steps involved include:

Data Preprocessing – Normalization of numerical features using MinMaxScaler.

Model Training – Training the autoencoder exclusively on non-failure data to learn normal operating patterns.

Anomaly Detection – Computing the reconstruction error and setting a threshold to classify anomalies.

Performance Evaluation – Comparing predictions against actual failures using a confusion matrix and recall score.

Results and Findings

The trained autoencoder successfully detected anomalies by setting an appropriate threshold for reconstruction error. The key results include:

Recall Score: 49.1% – The model correctly identified nearly half of actual machine failures.

False Positives: 232 instances – Some normal machine readings were misclassified as failures.

Confusion Matrix Analysis – The model correctly classified the majority of normal instances but still had some missed failures.

Feature Importance Analysis

Although autoencoders do not provide explicit feature importance, we used reconstruction errors and feature dropout tests to identify key contributors to machine failure. The most impactful features were:

Torque & Rotational Speed – Large fluctuations correlated with failures.

Process Temperature – Unusual temperature spikes were linked to breakdowns.

Power Consumption – Machines consuming excessive power showed higher failure probabilities.

Business Impact

Implementing this predictive maintenance model offers several advantages:

 Reduced Downtime – Early detection of failures prevents sudden breakdowns.

 Cost Savings – Lower maintenance and repair expenses.

 Improved Safety & Compliance – Reduces workplace hazards and regulatory risks.

 Extended Equipment Lifespan – Prevents excessive wear and tear.

 Optimized Resource Allocation – Enables data-driven maintenance scheduling.

Future Improvements

To further enhance model performance, the following improvements are recommended:

Hyperparameter Tuning – Adjusting neural network layers and loss functions for better reconstruction accuracy.

Hybrid Models – Combining autoencoders with traditional anomaly detection techniques (e.g., Isolation Forest, One-Class SVM).

Real-Time Monitoring System – Deploying the model as a real-time dashboard for live failure detection.

Conclusion

This project demonstrates the effectiveness of autoencoder-based anomaly detection in predicting machine failures. By leveraging sensor data, businesses can transition from reactive to predictive maintenance strategies, improving efficiency and reducing operational costs. Further refinements, such as integrating hybrid models and real-time monitoring, can make this solution even more robust for industrial applications

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
````

````python
import numpy as np

# Handling Outliers using IQR (Interquartile Range) Method

# Define a function to detect and handle outliers using the IQR method
def handle_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR  # Lower threshold
        upper_bound = Q3 + 1.5 * IQR  # Upper threshold

        # Replace outliers with the median value of the column
        median_value = df[col].median()
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median_value, df[col])

    return df

# Identify numerical columns for outlier treatment
numerical_cols = df.select_dtypes(include=['number']).columns

# Apply outlier handling
df_cleaned = handle_outliers_iqr(df, numerical_cols)

# Verify if outliers are handled by checking new IQR ranges
new_outlier_counts = {}
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    new_outlier_counts[col] = ((df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)).sum()

# Return the number of outliers detected before and after treatment
new_outlier_counts
````
