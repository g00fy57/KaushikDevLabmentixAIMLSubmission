## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

Null Hypothesis (H₀): There is no significant difference between the Open and Close prices.

Alternative Hypothesis (H₁): There is a significant difference between the Open and Close prices.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
````

````python
from scipy import stats

# Perform Paired t-test for Open vs. Close prices
t_stat, p_value_ttest = stats.ttest_rel(df["Open"], df["Close"])

# Display results
{
    "Test": "Paired t-test (Open vs. Close)",
    "t-statistic": t_stat,
    "p-value": p_value_ttest,
}

````

t-statistic: 0.221

p-value: 0.825

Conclusion: Since p > 0.05, we fail to reject the null hypothesis. This means there is no significant difference between Open and Close prices, suggesting they are statistically similar.

##### Which statistical test have you done to obtain P-Value?

Answer Here.

Paired t-test

##### Why did you choose the specific statistical test?

Answer Here.

 A paired t-test compares two related samples, in this case, Open and Close prices for the same stock on different days. Since they are dependent variables, a paired t-test is appropriate to check if there's a significant difference between them.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

Null Hypothesis (H₀): The variance of stock prices (High - Low) remains constant over time.

Alternative Hypothesis (H₁): The variance of stock prices changes significantly over time.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
````

````python
# Compute stock price volatility as (High - Low)
volatility = df["High"] - df["Low"]
mid = len(volatility) // 2  # Split dataset into two halves

# Perform Levene’s test for equality of variance
levene_stat, p_value_levene = stats.levene(volatility[:mid], volatility[mid:])

# Display results
{
    "Test": "Levene’s Test (Volatility over time)",
    "Levene-statistic": levene_stat,
    "p-value": p_value_levene,
}
````

Levene-statistic: 37.78
p-value: 4.84e-09 (very small)

Conclusion: Since p < 0.05, we reject the null hypothesis, meaning that stock price volatility (High - Low) has significantly changed over time. This suggests that market conditions have fluctuated.

##### Which statistical test have you done to obtain P-Value?

Answer Here.

Levene's Statistical Test

##### Why did you choose the specific statistical test?

Answer Here.

Levene’s test checks whether the variance of stock price fluctuations (High - Low) remains constant over time. Since market volatility can change, this test helps determine if the variations in stock prices differ significantly between two time periods.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

Null Hypothesis (H₀): There is no significant correlation between Open and Close prices.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
````

````python
# Perform Pearson Correlation Test for Open vs. Close prices
corr_coeff, p_value_corr = stats.pearsonr(df["Open"], df["Close"])

# Display results
{
    "Test": "Pearson Correlation (Open vs. Close)",
    "Correlation Coefficient": corr_coeff,
    "p-value": p_value_corr,
}
````

````python
import numpy as np
from sklearn.feature_selection import VarianceThreshold

# Step 1: Remove Highly Correlated Features (Threshold = 0.85)
correlation_matrix = df_final.corr()
high_corr_pairs = correlation_matrix.abs().where(np.triu(np.ones(correlation_matrix.shape), k=1) > 0.85)
drop_cols = [column for column in high_corr_pairs.columns if any(high_corr_pairs[column] > 0.85)]

# Drop highly correlated features
df_reduced = df_final.drop(columns=drop_cols)

# Step 2: Remove Low-Variance Features
selector = VarianceThreshold(threshold=0.01)  # Removing near-constant features
df_selected = df_reduced.loc[:, selector.fit(df_reduced).get_support()]

#Step 3: Remove Price Change column as it contains NaN Value
df_selected.drop(columns=["Price_Change"], inplace=True)

# Display selected features after statistical selection
df_selected.head(), df_selected.columns.tolist()

````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
````

````python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge

# Initialize Ridge Regression model (L2 Regularization) with different alpha values
alpha_values = [0.01, 0.1, 1, 10, 100]
best_alpha = None
best_score = -float("inf")

# Perform cross-validation for each alpha value
for alpha in alpha_values:
    ridge_model = Ridge(alpha=alpha)
    scores = cross_val_score(ridge_model, X_train, np.log1p(y_train), cv=5, scoring="r2")  # 5-Fold CV

    avg_r2 = scores.mean()  # Average R² score across folds

    if avg_r2 > best_score:
        best_score = avg_r2
        best_alpha = alpha

# Train Ridge Regression with the best alpha
ridge_best = Ridge(alpha=best_alpha)
ridge_best.fit(X_train, np.log1p(y_train))

# Predict on test set
y_pred_log_ridge = ridge_best.predict(X_test)
y_pred_ridge = np.expm1(y_pred_log_ridge)

# Evaluate final model
r2_ridge = r2_score(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))

best_alpha, r2_ridge, rmse_ridge
````
